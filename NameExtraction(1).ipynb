```python
import requests
from bs4 import BeautifulSoup
import probablepeople as pp
import time

def checkPersonOnline(name):

    page = requests.get("https://en.wikipedia.org/wiki/" + name)
    soup = BeautifulSoup(page.content, 'html.parser')

    personHeaders = ['career', 'biography', 'early life', 'death', 'marriages', 'marriage', 'life']


    headersString  = ""
    for headlines in soup.find_all("h2"):
        headersString  = headersString + headlines.text.strip().replace('[edit]', '').lower()+" "

    for headlines in soup.find_all("h3"):
        headersString  = headersString + headlines.text.strip().replace('[edit]', '').lower()+" "
    

    for item in personHeaders:
        if item in headersString:
            return 1
    return 0
```

```python
def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)

def isName(array):
    if (len(array) < 2):
        return 0
    
    for set in array:
        if(set[1] not in ['Surname', 'GivenName']):
            return 0
    return 1
```
